## 1. Part 1 - Charts

#### Histogram

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better visualization
plt.figure(figsize=(12, 8))  # Made figure slightly larger to accommodate text

# Create histogram
posts_df['Score'].hist(bins=50)

# Add main title and axis labels
plt.title('Distribution of Post Scores')
plt.xlabel('Score')
plt.ylabel('Frequency (log scale)')
plt.yscale('log')
plt.grid(True, alpha=0.3)

description = """\
This histogram shows the distribution of post scores from the Apple Stack Exchange forum.
• Most posts have scores between 0-20
• The distribution is right-skewed
• Very few posts achieve scores above 40
• The y-axis uses logarithmic scale for better visualization\
"""

# Position text in the upper right portion of the plot
plt.text(22, 1200, description,
         bbox=dict(facecolor='white', alpha=0.8),
         verticalalignment='top',
         fontsize=10)

# Adjust layout to prevent text cutoff
plt.tight_layout()
plt.show()
````

#### Box-and-whiskers plot

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract year from CreationDate
users_df['JoinYear'] = pd.to_datetime(users_df['CreationDate']).dt.year

plt.figure(figsize=(12, 8))
sns.boxplot(x='JoinYear', y='Reputation', data=users_df)
plt.title('User Reputation Distribution by Join Year')
plt.xlabel('Join Year')
plt.ylabel('Reputation (log scale)')
plt.yscale('log')
plt.xticks(rotation=45)

# Add descriptive text
description = """\
This box plot shows the distribution of user reputation scores by join year:
• Each box represents users who joined in that specific year
• The line in each box is the median reputation
• Boxes show the interquartile range (IQR: 25th to 75th percentile)
• Whiskers extend to show the rest of the distribution
• Points beyond whiskers represent outliers (users with exceptional reputation)
• Y-axis uses logarithmic scale due to wide reputation range
• Earlier users tend to have higher reputation scores\
"""

plt.text(0.465, 0.975, description,
         transform=plt.gca().transAxes,
         bbox=dict(facecolor='white', alpha=0.8),
         verticalalignment='top',
         fontsize=10)

plt.tight_layout()
plt.show()
````

#### Heatmap


````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Convert CreationDate to datetime if not already
posts_df['CreationDate'] = pd.to_datetime(posts_df['CreationDate'])

# Extract day of week and hour
posts_df['DayOfWeek'] = posts_df['CreationDate'].dt.day_name()
posts_df['Hour'] = posts_df['CreationDate'].dt.hour

# Create pivot table for heatmap
activity_pivot = pd.crosstab(posts_df['DayOfWeek'], posts_df['Hour'])

# Reorder days to start with Monday
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
activity_pivot = activity_pivot.reindex(day_order)

plt.figure(figsize=(15, 8))
sns.heatmap(activity_pivot, cmap='YlOrRd', annot=True, fmt='d', cbar_kws={'label': 'Number of Posts'})

plt.title('Post Activity Heatmap by Day and Hour')
plt.xlabel('Hour of Day (24h format)')
plt.ylabel('Day of Week')

# Add descriptive text
description = """This heatmap shows posting activity patterns:
• X-axis shows hour of day (0-23)
• Y-axis shows day of week
• Color intensity indicates post volume
• Numbers show exact post counts
• Darker colors = higher activity
• Helps identify peak posting times"""

plt.text(0, -0.1, description,
         transform=plt.gca().transAxes,
         bbox=dict(facecolor='white', alpha=0.8),
         verticalalignment='top',
         fontsize=10)

plt.tight_layout()
plt.show()
````

#### Bubble chart

````python
plt.figure(figsize=(12, 8))

# Calculate post count per user
post_counts = posts_df['OwnerUserId'].value_counts()
top_users = post_counts.head(20)  # Get top 20 users

# Get user data for these users
top_user_data = users_df[users_df['Id'].isin(top_users.index)]

# Create bubble chart
plt.scatter(top_user_data['Reputation'], 
           top_user_data['UpVotes'],
           s=top_users.values * 20,  # Size based on post count (multiplied for visibility)
           alpha=0.6)

plt.xlabel('Reputation')
plt.ylabel('UpVotes Given')
plt.title('Top Users: Reputation vs UpVotes (Bubble Size = Post Count)')
plt.yscale('log')
plt.xscale('log')

# Add descriptive text
description = """This bubble chart shows relationships between user metrics:
• X-axis: User reputation (log scale)
• Y-axis: Number of upvotes given (log scale)
• Bubble size: Number of posts made
• Each bubble represents one user
• Shows how activity relates to reputation
• Larger bubbles indicate more active users"""

plt.text(0.03, 0.95, description,
         transform=plt.gca().transAxes,
         bbox=dict(facecolor='white', alpha=0.8),
         verticalalignment='top',
         fontsize=10)

plt.tight_layout()
plt.show()
````

#### Code

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure 'CreationDate' is in datetime format
posts_df['CreationDate'] = pd.to_datetime(posts_df['CreationDate'])

# Extract Year and Month for grouping
posts_df['YearMonth'] = posts_df['CreationDate'].dt.to_period('M')

# Function to extract tags from the '<tag1><tag2>' format
def extract_tags(tag_str):
    if pd.isna(tag_str):
        return []
    return tag_str.strip('<>').split('><')

# Create a list of tags for each post
posts_df['TagList'] = posts_df['Tags'].apply(extract_tags)

# Explode the dataframe so each tag has its own row
exploded_tags = posts_df.explode('TagList')

# Identify top 5 tags overall
top_tags = exploded_tags['TagList'].value_counts().head(5).index.tolist()

# Filter the dataset for only the top tags
top_tags_df = exploded_tags[exploded_tags['TagList'].isin(top_tags)]

# Group by 'YearMonth' and 'TagList' to get counts
tag_trends = top_tags_df.groupby(['YearMonth', 'TagList']).size().reset_index(name='Counts')

# Pivot the data for plotting
tag_trends_pivot = tag_trends.pivot(index='YearMonth', columns='TagList', values='Counts').fillna(0)

# Sort the index
tag_trends_pivot = tag_trends_pivot.sort_index()

# Plotting
plt.figure(figsize=(14, 8))

# Plot stacked area chart
tag_trends_pivot.plot(kind='area', stacked=True, colormap='Spectral', alpha=0.8, ax=plt.gca())

plt.title('Trend of Top 5 Tags Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Number of Posts')
plt.legend(title='Tags', loc='upper left')

# Add descriptive text
description = """This stacked area chart shows the posting trends of the top 5 most used tags over time:
• X-axis represents time from earliest to latest posts
• Y-axis represents the total number of posts
• Each colored area represents a tag, showing how its usage changes over time
• Highlights the rise or decline in popularity of certain topics within the community
• Provides insight into shifting interests and discussions"""

plt.text(0.5, -0.15, description,
         transform=plt.gca().transAxes,
         bbox=dict(facecolor='white', alpha=0.8),
         verticalalignment='top',
         horizontalalignment='center',
         fontsize=10)

plt.tight_layout()
plt.show()
````
